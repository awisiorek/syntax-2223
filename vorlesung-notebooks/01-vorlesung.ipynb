## 2. NLTK: Syntaxanalyse mit formalen Grammatiken

````python
import nltk
````

````python
# Inputsatz
sent = 'I shot an elephant in my pajamas'.split()
print(sent)
````

---
### 2.1 CFG (Kontextfreie Grammatik)
- Konstituenzrelationen zwischen W√∂rtern und syntaktischen Einheiten (Phrasen etc.)
- Strukturinformationen in Knoten

````python
#CFG Grammatik:
grammar = nltk.CFG.fromstring("""
    S -> NP VP
    PP -> P NP
    NP -> Det N | Det N PP | Pron
    VP -> V NP | VP PP
    Pron -> 'I'
    Det -> 'an' | 'my'
    N -> 'elephant' | 'pajamas'
    V -> 'shot'
    P -> 'in'
    """)
````

````python
#Parsing = √úberpr√ºfung Inputsatz auf Wohlgeformtheit gem√§√ü Grammatik:
parser = nltk.ChartParser(grammar,trace=0)

for tree in parser.parse(sent):
    print(tree)
    tree.pretty_print(unicodelines=True)
    
# Konstituentenstruktur:   
````

---

### 2.2 Dependenzgrammatik
- Abh√§ngigkeitsrelationen zwischen W√∂rtern eines Satzes (Head-Dependent-Relationen)
- Strukturinformationen in Kanten (ggf. gelabelt mit syntaktischen Funktionen)

````python
#Dependency Grammar (unlabeled):
grammar = nltk.DependencyGrammar.fromstring("""
    'shot' -> 'I' | 'elephant' | 'in'
    'elephant' -> 'an' | 'in'
    'in' -> 'pajamas'
    'pajamas' -> 'my'
    """)
````

````python
#Parsing mit Dependenzgrammatik:
parser = nltk.ProjectiveDependencyParser(grammar)

for tree in parser.parse(sent):
    print(tree)
    tree.pretty_print(unicodelines=True)
    
# Dependenzstruktur:
````

---
### 2.3 Partielles Parsing mit regul√§rer Grammatik (Chunking)

````python
# partielle, flache RegExp-Grammatik:
grammar = r"""
    NP: {<DET>?<ADJ>*<N>} 
        {<PRON>}  
    PP: {<P>}
"""

sent = [("I", "PRON"), ("shot", "V"), ("an", "DET"), ("elephant", "N"), 
        ("in", "P"), ("my", "DET"), ("pajamas", "N")]

#Parsing mit RegExp-Grammatik:
parser = nltk.RegexpParser(grammar)

tree = parser.parse(sent) 
print(tree)
tree.pretty_print(unicodelines=True)
````

## 3. CFG-Erweiterungen

### 3.1 PCFG (Probabilistische CFG)

- Viterbi-Parser findet Ableitung mit der h√∂chsten Wahrscheinlichkeit (erm√∂glicht Disambiguierung)
- Wahrscheinlichkeiten f√ºr Regeln werden √ºblicherweise aus syntaktisch annotiertem Korpus (Treebank) gelernt

````python
# Inputsatz
sent = 'I shot an elephant in my pajamas'.split()
print(sent)
````

````python
# PCFG-Grammatik:
grammar1 = nltk.PCFG.fromstring("""
    S -> NP VP [1.0]
    PP -> P NP [1.0]
    NP -> Det N  [0.8] | Det N PP  [0.1] | Pron  [0.1]
    VP -> V NP [0.8] | VP PP [0.2]
    Pron -> 'I' [1.0]
    Det -> 'an' [0.7] | 'my' [0.3]
    N -> 'elephant' [0.5] | 'pajamas' [0.5]
    V -> 'shot' [1.0]
    P -> 'in' [1.0]
    """)
````

````python
parser = nltk.ViterbiParser(grammar1)

for tree in parser.parse(sent):
    print(tree)
    tree.pretty_print(unicodelines=True)

#PP-Attachment an VP (h√∂here Wahrscheinlichkeit dieses Syntaxbaums) 
# Disambiguierung! 
````

````python
# Grammatik mit h√∂herer Gewichtung f√ºr Det N PP - Regel:

grammar2 = nltk.PCFG.fromstring("""
    S -> NP VP [1.0]
    PP -> P NP [1.0]
    NP -> Det N  [0.7] | Det N PP  [0.2] | Pron  [0.1] 
    VP -> V NP [0.8] | VP PP [0.2]
    Pron -> 'I' [1.0]
    Det -> 'an' [0.7] | 'my' [0.3]
    N -> 'elephant' [0.5] | 'pajamas' [0.5]
    V -> 'shot' [1.0]
    P -> 'in' [1.0]
    """)
````

````python
parser = nltk.ViterbiParser(grammar2)

for tree in parser.parse(sent):
    print(tree)
    tree.pretty_print(unicodelines=True)
    
#PP-Attachment an NP
````

---
### 3.2 *Feature-based* CFG

- Erweiterung CFG um grammatische Merkmale verhindert √úbergenerierung

````python
# Ausgangspunkt: mehr Regeln in CFG (f√ºhrt zu √úbergenerierung)
grammar = nltk.CFG.fromstring("""
    S -> NP VP
    PP -> P NP
    NP -> Det N | Det N PP | Pron
    VP -> V NP | VP PP
    Pron -> 'I' | 'He'
    Det -> 'an' | 'my'
    N -> 'elephant' | 'pajamas'
    V -> 'shot' |¬†'shoot' | 'shoots'
    P -> 'in'
    """)

parser = nltk.ChartParser(grammar,trace=0)
````

````python
sent = 'He shoots an elephant'.split()

for tree in parser.parse(sent):
    print(tree)
    tree.pretty_print(unicodelines=True)
````

````python
#Negativbeispiel (ungrammatischer Satz):
sent = 'I shoots an elephant'.split()

for tree in parser.parse(sent):
    print(tree)
    tree.pretty_print(unicodelines=True)

#√úbergenerierung durch neue CFG-Regeln!
````

````python
# FCFG = Feature-Based Grammar (PER-Agreement-Constraint):
gramstring = r"""
% start S
    S -> NP[PER=?x] VP[PER=?x]
    PP -> P NP
    NP[PER=?x] -> Det N | Det N PP | Pron[PER=?x]
    VP[PER=?x] -> V[PER=?x] NP | VP PP
    Pron[PER=1] -> 'I'
    Pron[PER=3] -> 'He'
    Det -> 'an' | 'my'
    N -> 'elephant' | 'pajamas'
    V[PER=1] -> 'shoot'
    V[PER=3] -> 'shoots'
    P -> 'in'
"""

grammar = nltk.grammar.FeatureGrammar.fromstring(gramstring)
parser = nltk.parse.FeatureChartParser(grammar, trace=1)
````

````python
#Positivbeispiel (grammatischer Satz):
sent = 'He shoots an elephant'.split()

for tree in parser.parse(sent):
    print(tree)
    tree = nltk.Tree.fromstring(str(tree).replace(", ",","))
    tree.pretty_print(unicodelines=True)
    
#Parse gefunden (siehe Trace-Output!)
````

````python
#Negativbeispiel (ungrammatischer Satz):
sent = 'I shoots an elephant'.split()

for tree in parser.parse(sent):
    print(tree)

#kein Parse gefunden (siehe Trace-Output!)
# PER-Feature-Constraint verhindert √úbergenerierung!
````
